```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

from src.functions import array_to_vector, quadratic, rosenbrook, sty_tang, rastrigin
```

```python
def quadratic2d(x, y):
    v = array_to_vector([x, y])
    return quadratic(v)
```

```python
def rosenbrook2d(x, y):
    v = array_to_vector([x, y])
    return rosenbrook(v)
```

```python
def sty_tang2d(x, y):
    v = array_to_vector([x, y])
    return sty_tang(v)
```

```python
def rastrigin2d(x, y):
    v = array_to_vector([x, y])
    return rastrigin(v)
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
function = quadratic2d(X, Y)
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, function, cmap='viridis')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_zlabel('$f$')
ax1.set_title('Gráfico da quadrática')
```

```python
fig = plt.figure(figsize=(6, 5))
cp = plt.contourf(X, Y, function, cmap='viridis')
plt.colorbar(cp)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Curvas de nível da quadrática')
plt.show()
```

```python
x = np.linspace(-2, 2, 1000)
y = np.linspace(-1, 3, 1000)
X, Y = np.meshgrid(x, y)
function = rosenbrook2d(X, Y)
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, function, norm=colors.LogNorm(vmin=10 ** (-4), vmax=1000), cmap='viridis')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_zlabel('$f$')
ax1.set_title('Gráfico da Rosenbrook')
```
